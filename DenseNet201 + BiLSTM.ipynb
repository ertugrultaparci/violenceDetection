### DenseNet201 + Bidirectional LSTM

```python
#Import model libraries:
from keras.models import Model, Sequential
from keras.layers import LSTM, Bidirectional

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

from tensorflow.keras.applications.densenet import DenseNet201
from dataPreparation import *
```

```python
img_size = 224
img_size_flat = img_size * img_size * num_channels
num_classes = 2
_num_files_train = 1
_images_per_file = 20
_num_images_train = _num_files_train * _images_per_file
video_exts = ".avi"
```

```python
data, target = process_alldata_training('densenet_feature_extraction.h5')
data_test, target_test = process_alldata_test('densenet_feature_extraction_validation.h5')
```

```python
from keras import backend as K
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten,Conv2D, MaxPooling2D,LSTM,Bidirectional
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.callbacks import TensorBoard

NAME="DenseNet"

tensorboard=TensorBoard(log_dir='logs/{}'.format(NAME))
model=Sequential()
model.add(Bidirectional(LSTM(64, input_shape=(20, 4096))))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(2,activation='softmax'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history = model.fit(np.array(data[0:600]), np.array(target[0:600]),epochs=500,
                    validation_data=(np.array(data[600:]), np.array(target[600:])),
                    batch_size=500,callbacks=[tensorboard])
model.save("DenseNet201.h5")
```

```python
# MLP for Pima Indians Dataset Serialize to JSON and HDF5
from tensorflow.keras.models import Sequential, model_from_json
# serialize model to JSON
model_json = model.to_json()
with open("model_densenetbilstm.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("densenetbilstm.h5")
print("Saved model to disk")
 
# later...
 
# load json and create model
json_file = open('model_densenetbilstm.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("densenetbilstm.h5")
print("Loaded model from disk")
```

```python
loaded_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
result = loaded_model.evaluate(np.array(data_test), np.array(target_test))
```
